```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/data_cleaned_2021.csv')
df_useful = df[['Job Title', 'Rating', 'Location', 'Headquarters', 'Size', 'company_txt', 'Industry', 'Lower Salary', 'Upper Salary', 'Avg Salary(K)', 'job_title_sim', 'seniority_by_title', 'Python', 'spark', 'aws', 'sql', 'keras', 'pytorch', 'scikit', 'tensor', 'hadoop', 'flink', 'mongo']]
df_useful
```

```python
df_useful.replace(-1, None, inplace=True)
df_useful.replace('-1', None, inplace=True)
```

```python
def transform_size(company_size):
    if '10000+' in company_size:
        return 9
    elif 'unknown' in company_size:
        return 2
    else:
        size = int(company_size.split('-')[0])
        if size == 1:
            return 3
        elif size == 51:
            return 4
        elif size == 201:
            return 5
        elif size == 501:
            return 6
        elif size == 1001:
            return 7
        elif size == 5001:
            return 8
```

```python
df_useful.loc[:, 'Size'] = df_useful['Size'].apply(transform_size)
df_useful.head()
```

```python
def fix_missing_industry(row):
    if 'Kronos Bi' in row['company_txt']:
        row['company_txt'] = 'Kronos Bio'
        row['Industry'] = 'Biotech & Pharmaceuticals'
        row['Rating'] = df_useful[df_useful['Industry'] == 'Biotech & Pharmaceuticals']['Rating'].mean()
    elif 'Persivia' in row['company_txt']:
        row['Industry'] = 'Health Care Services & Hospitals'
    elif 'Monte Rosa Therapeutic' in row['company_txt']:
        row['Industry'] = 'Biotech & Pharmaceuticals'
        row['Headquarters'] = 'Boston, MA'
        row['Rating'] = df_useful[df_useful['Industry'] == 'Biotech & Pharmaceuticals']['Rating'].mean()
    elif 'ALI' in row['company_txt']:
        row['Industry'] = 'IT Services'
        row['Rating'] = df_useful[df_useful['Industry'] == 'IT Services']['Rating'].mean()
    elif 'Mus' in row['company_txt']:
        row['Industry'] = 'IT Services'
        row['Rating'] = df_useful[df_useful['Industry'] == 'IT Services']['Rating'].mean()
    else:
        pass
    return row
    
```

```python
df_useful = df_useful.apply(fix_missing_industry, axis = 1)
df_useful.head()
```

```python
# assess if there's any more null values
df_useful[df_useful.isnull().any(axis=1)]
```

```python
it_avg = df_useful[df_useful['Industry'] == 'IT Services']['Rating'].mean()
def imputeRating(row):
    if np.isnan(row['Rating']):
        row['Rating'] = it_avg
    else:
        pass
    return row
df_useful = df_useful.apply(imputeRating, axis = 1)
df_useful.head()
```

```python
df_useful[df_useful.isnull().any(axis=1)]
```

```python
def clean_job_title(row):
    if 'Data' in row['Job Title'] and 'Engineer' in row['Job Title']:
        row['job_title_sim'] = 'data engineer'
    elif 'Data' in row['Job Title'] and 'Management' in row['Job Title']:
        row['job_title_sim'] = 'data engineer'
    elif 'Enterprise' in row['Job Title'] and 'Architect' in row['Job Title']:
        row['job_title_sim'] = 'data engineer'
    elif 'Operation' in row['Job Title']:
        row['job_title_sim'] = 'analyst'
    elif 'Professional' in row['Job Title'] and 'Data' in row['Job Title']:
        row['job_title_sim'] = 'analyst'
    else:
        pass
    return row
```
